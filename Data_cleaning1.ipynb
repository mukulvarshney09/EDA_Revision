## Data is clean, let's make some visual presentations and remarks

### Histogram

```python
d1["TV"]
```

```python
plt.hist(x = d1["TV"])
plt.xlabel("TV Sales")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(x = d1["Radio"])
plt.xlabel("Radio Sales")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(x = d1["Newspaper"])
plt.xlabel("Newspaper Sales")
plt.ylabel("Density")
plt.show()
```

### DIST Plot

```python
sns.distplot(x = d1["Newspaper"], kde = True)
plt.xlabel("Newspaper Sales")
plt.ylabel("Density")
plt.show()
```

```python
sns.distplot(x = d1["TV"], kde = True)
plt.xlabel("TV Sales")
plt.ylabel("Density")
plt.show()
```

```python
sns.distplot(x = d1["Radio"], kde = False)
plt.xlabel("Radio Sales")
plt.ylabel("Density")
plt.show()
```

```python
sns.kdeplot(data = d1, x = d1["TV"], fill = True)
plt.xlabel("TV Sales")
plt.ylabel("Density")
plt.show()
```

### Line Plot

```python
x = d1["Newspaper"]
y = d1["Sales"]
plt.plot(x,y)
plt.xlabel("Newspaper")
plt.ylabel("Sales")
plt.title("Newspaper vs Sales")
plt.show()
```

```python
x = d1["TV"]
y = d1["Sales"]
plt.plot(x,y)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("TV vs Sales")
plt.show()
```

```python
x = d1["Radio"]
y = d1["Sales"]
plt.plot(x,y)
plt.xlabel("Radio")
plt.ylabel("Sales")
plt.title("Radio vs Sales")
plt.show()
```
