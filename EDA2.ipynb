```python
#categorical columns graphs
plt.figure(figsize=(20,15))
plt.suptitle("Univariate analysis of categorical data", fontsize = 20, alpha = 1.0)
columns = ["Type", "Content Rating"]
for i in range(0, len(columns)):
  plt.subplot(2,2,i+1)
  sns.countplot(x = df[columns[i]], palette = "Set2")
  plt.xlabel(columns[i])
  plt.tight_layout()
  plt.xticks(rotation=0)

```

```python
plt.figure(figsize=(20,15))
plt.suptitle("Univariate analysis of categorical data", fontsize = 20, alpha = 1.0)
columns = ["Type", "Content Rating"]
for i in range(0, len(columns)):
  plt.subplot(2,2,i+1)
  sns.countplot(x = df[columns[i]], palette = "Set3")
  plt.xlabel(columns[i])
  plt.tight_layout()
  plt.xticks(rotation=30)
```
